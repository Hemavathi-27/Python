```python
#nested conditional statements
num=int(input("Enter a number to check whether it is odd or even"))
if num>0:
    print("the number is positive")
    if num%2==0:
        print("it is even number")
    else:
        print("it is odd number")
else:
    print("it is a negative number")



```

```python
##use cases
##dtermine if a year is a leap year using nested conditional statements
year=int(input("Enter a year to check whether it is a leap year or not"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"is a leap year")
        else:
            print(year,"is not a leap year")
    else:
        print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```
